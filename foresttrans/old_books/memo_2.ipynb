```python
def get_median_v2(input_num):
    """ get_median_v2 """
    input_num = sorted(input_num)  #주어진 리스트 정렬하기
    center_Index=len(input_num)//2 #center_Index 값 할당하기
    return (input_num[center_Index]+input_num[-center_Index -1])/2


def get_count_by_query(query, min_cnt=0, verbose=0):
    datas = []
    rows = !echo "{query}"|sqlite3 ../db/metrics.db
    if verbose > 1:
        print(row)
    for row in rows:
        if len(row) > 0 and row[0].isdigit():
            cols = row.split('|')
            if len(cols) == 3:
                cnts = [int(i) for i in cols[2].split(',')]
                count = 0
                for cnt in cnts:
                    if cnt >=3:
                        count += 1
                if count >= min_cnt:
                    datas.append([cols[0], cols[1], cnts, count, ])
    return datas


def get_count(random_states, version, min_cnt=0, verbose=0):
    """ get_count """
    datas = []
    if len(random_states) > 0:
        for random_state in random_states:
            t_version = random_state[1]
            query = f'select random_state, version, matched_cnts from rndforest where version=\\"{version}\\" and random_state={random_state[0]}'
            result = get_count_by_query(query=query, min_cnt=min_cnt, verbose=verbose)
            if len(result) > 0:
                datas.append(result[0])
    else:
        query = f'select random_state, version, matched_cnts from rndforest where version=\\"{version}\\"'
        datas = get_count_by_query(query=query, min_cnt=min_cnt, verbose=verbose)
    return datas


def get_count_with_start_width(start, data_length, version, min_cnt=0, verbose=0):
    if start > 0:
        random_states = [(i, version) for i in range(start, start+data_length)]
    else:
        random_states = []
    return get_count(random_states=random_states, version=version, min_cnt=min_cnt, verbose=verbose)


print('current', datetime.now())
```
