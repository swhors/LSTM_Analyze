````python
def print_near_metric(matched_pos, left_depth=2, right_depth=2):
    find_pos = 0
    for result in sorted_results2:
        if result[0] == matched_pos:
            break
        else:
            find_pos += 1
    from_pos = find_pos - left_depth
    to_pos = find_pos + right_depth
    cur_pos = from_pos

    while cur_pos <= to_pos:
        print(sorted_results2[cur_pos])
        cur_pos += 1
````
