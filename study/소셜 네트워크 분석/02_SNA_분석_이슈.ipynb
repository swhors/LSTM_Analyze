## 02. SNA 분석 이슈

### 2-1. SNA 분석에 관한 신문기사(2)

    "특허 공동출원 분석해보니 .... 현대車 '지식영향력', 삼성전자 '지식 정보통제력'이 컸다", 워크투데이, 2024-04-04.
    지난 10년간 한국의 특허 네트워크는 더 촘촘해지고 군집화·지역화된 것으로 나타났다.
    이러한 연구협력 네트워크는 특히 바이오, 화학, 자동차 제조업, 연구개발서비스업과 같이 첨단산업이나 대규모 설비투자가 필요한 중후장대 산업에서 상대적으로 더 촘촘하게 형성돼 있었던 것으로 파악됐다.
    한국지식재산연구원 특허통계센터는 최근 10년간(2013~2023년9월) 국내에 출원된 전체 특허 중 출원인(개인 제외)이 복수인 약 13만 개 특허의 공동출원 정보를 바탕으로 우리나라의 연구 협력 네트워크를 분석한 결과 이같이 나타났다고 밝혔다.
    분석결과에 따르면 특허 공동출원을 통해 형성된 기관 간 연결 수는 2013~2017년 2만7,172건이었으나 2018~2022년의 경우 3만590건으로 늘어났다. 같은 기간 동일 권역 내에서 네트워크가 형성된 비중도 52.2%에서 65.1%로 높아졌다.
    보고서는 특허 네트워크 내에서 누가 더 중요한 혁신 주체인지를 의미하는 네트워크 중심성도 분석했다.
    그 결과 연결중심성(해당 군집에서의 영향력)은 현대자동차, 근접중심성(다른 기관과 가장 가까이 있는 정도)은 서울대학교, 매개중심성(기관 간 연결을 가장 효율적으로 해주는 경로 위에 있는 정도)은 삼성전자가 10년간 꾸준히 1위를 유지하고 있는 것으로 나타났다.
    쉽게 말해 현대자동차는 타 주체에 미치는 지식 영향력이 가장 크고, 서울대학교는 타 주체와 가장 가깝게 있고, 삼성전자는 지식 확산에 있어서 정보통제력이 컸다고 지식재산연구원은 설명했다.
    임소진 한국지식재산연구원 특허통계센터 박사는 “특허의 공동출원 정보는 소셜 네트워크 분석(SNA)을 통해 지식의 확산과 흐름을 파악할 수 있는 매우 유용한 자료”라고 말했다.
    이인수 특허청 산업재산정보국장은 “반도체, 이차전지 등 우리나라의 전략기술 분야에 형성돼 있는 특허 네트워크의 다면적 구조와 각 주체의 역할까지 종합적으로 고려하는 국가 R&D 전략을 수립할 필요가 있다”고 말했다.

### 2-2. SNA 분석에 관한 학술연구

- "코로나 팬데믹에 따른 항만물류 네트워크 변화 분석 연구" , 손유미·김화영, 한국항만경제학회지, 2023
  - 본 연구는 코로나19 팬데믹 전후의 항만 물류 변화의 영향을 살펴본다. 구체적으로 본 연구는 한국의 컨테이너 항만 네트워크의 변화를 분석하는 데 중점을 둔다. 나아가 본 연구는 중심성 지수(중심성 정도, 근접 중심성, 매개 중심성)와 같은 네트워크 분석 방법을 사용하여 코로나19 팬데믹 전후의 컨테이너 항만 네트워크에서 항만의 영향을 조사하여 2018년과 2021년 사이에 네트워크의 구조와 속성의 변화를 파악한다.
  - 본 연구에서는 한국의 5대 항만인 부산, 광양, 인천, 울산, 평택당진의 컨테이너 항만 네트워크의 변화를 분석한다. 그 결과, 부산항의 경우 싱가포르항이 중요한 역할을 하는 반면, 부산항은 광양, 인천, 울산항에서 핵심적인 역할을 한다. 광양항의 경우 말레이시아의 포트 켈랑은 말레이시아 정부의 팬데믹 극복 정책의 결과로 점점 더 영향력을 가지게 되었다. 인천항에서는 일본 항만이 자국 항만과 인천항 네트워크의 항만 사이에서 중개자 역할을 하고 있다. 울산항의 경우 한국 항만의 영향력이 높고, 평택-당진항의 경우 동남아시아 항만이 항만 사이의 중개자 역할을 하고 있다.
  - 본 연구는 한국의 항만 물류 네트워크의 변화를 분석함으로써 향후 COVID-19 팬데믹과 같은 항만 물류의 변화를 일으키는 불확실성 상황에 대응할 때 참고점으로 활용할 수 있다.
